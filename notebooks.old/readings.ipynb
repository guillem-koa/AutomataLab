# Readings
* Reads .xlsx readings located in a PLATE folder
* Grabs essential information and writes it in xlsx file 'outputs/readings.xlsx'

```python
from getexperimentdata import *
import numpy as np
import pandas as pd
import os
```

```python
experiments_dir = '/Users/guillemcobos/Library/CloudStorage/GoogleDrive-guillem.cobos@koabiotech.com/.shortcut-targets-by-id/1UQ_YnLRmoAqhCue-qXYNP9QhYorsiAxS/KOA Biotech/01. AUTOMATA LAB/EXPERIMENTS'
experiments_paths_all = sorted(os.listdir(experiments_dir))
experiments_paths = [os.path.join(experiments_dir, dir) for dir in experiments_paths_all[1:-2]]
experiments_paths = [dir for dir in experiments_paths if 'WATER' in dir or 'PAT' in dir]
#experiments_paths = experiments_paths[-15:] # select the folders you want to update
```

```python
plates_paths = []
for dir in experiments_paths:
    if os.path.exists(os.path.join(dir, 'PLATE1')): plates_paths.append(os.path.join(dir, 'PLATE1'))
    if os.path.exists(os.path.join(dir, 'PLATE2')): plates_paths.append(os.path.join(dir, 'PLATE2'))
```

```python
for plate_dir in plates_paths:
    print(plate_dir)
    os.chdir(plate_dir)
    # Get plate_distribution filepath
    plate_distribution_path = [file for file in os.listdir() if 'distribution' in file][0] if len([file for file in os.listdir() if 'distribution' in file])!=0 else ''
    readings_path = [file for file in os.listdir() if '.xlsx' in file and 'distribution' not in file][0] if len([file for file in os.listdir() if '.xlsx' in file and 'distribution' not in file])!=0 else ''

    if len(readings_path)!=0 and len(plate_distribution_path)!=0:
        plate_distribution_vars, sensors = get_plate_distribution_vars(plate_distribution_path)
        plate_distribution_vars_df = pd.DataFrame(plate_distribution_vars)
        
        try: 
            readings = pd.read_excel(readings_path)
            ABS, GFP = readings.iloc[2:10,3:14].values.astype(float), readings.iloc[23:31,3:14].values.astype(int)
            values_array = np.column_stack((ABS, GFP))
            values_column_names = [receptor + '_abs' for receptor in sensors] + [receptor + '_gfp' for receptor in sensors]
            values_df = pd.DataFrame(values_array, columns = values_column_names)
            readings_df = pd.concat([plate_distribution_vars_df,values_df], axis=1)

            folder_path = plate_dir + '/outputs'
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)
                
            readings_df.to_excel(os.path.join(plate_dir, 'outputs', 'readings.xlsx'), index=False)
            print(readings_df)

        except ValueError:
            print("Damaged readings xlsx file!")
```
